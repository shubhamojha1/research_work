```python
import pandas as pd
import numpy as np
from dateutil import parser
```

```python
df=pd.read_csv('dfas.csv')
```

```python
df
```

```python
cols_to_drop = [
    'Unnamed: 0',
    'job_name',
    'inst_id',
    'user',
    'group',
    'start_time_j',
     'end_time_j',
    'start_time_i',
     'end_time_i',
     'status_j',
     'status_i',
]
```

```python
df.drop(cols_to_drop, axis=1, inplace=True)
```

```python
numerical_cols = [x for x in list(df.columns) if str(df[x].dtype)!="object"]
categorical_cols = [x for x in df.columns if x not in numerical_cols]
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
```

```python
df['workload'].fillna('NA', inplace=True)
```

```python
df.isna().sum()
```

```python
df['start_date'] = pd.to_datetime(df['start_date'])
```

```python
def convert_to_unix_epoch(timestamp_str):
    timestamp = parser.parse(timestamp_str)
    unix_epoch_time = timestamp.timestamp()
    return unix_epoch_time
```

```python
# Apply the conversion function to the entire column
# df['timestamp'] = df['start_date'].apply(convert_to_unix_epoch)
```

```python
df['start_date'] = df['start_date'].dt.tz_localize(None)
```

```python
df['timestamp'] = (df['start_date'] - pd.Timestamp('1970-01-01 00:00:00')) // pd.Timedelta(seconds=1)
```

```python
df.set_index('timestamp', inplace=True)
```

```python
df.sort_index(inplace=True)
```

```python
df.reset_index(inplace=True)
```
